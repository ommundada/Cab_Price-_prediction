```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("updated_dataset.csv")
df.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures


selected_columns = ['surge_multiplier','distance']

degree = 2


poly_features = PolynomialFeatures(degree=degree)


poly_transform = poly_features.fit_transform(df[selected_columns])


poly_df = pd.DataFrame(poly_transform, columns=poly_features.get_feature_names(selected_columns))


df_poly = pd.concat([df.drop(columns=selected_columns), poly_df], axis=1)

df_poly
```

```python
df_poly.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
x = df_poly.drop('price',axis=1)
y = df_poly['price']
print(x.shape,y.shape)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42015)
print(xtrain.shape,ytrain.shape,xtest.shape,ytest.shape)
```

```python
model.fit(xtrain,ytrain)
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import r2_score,mean_absolute_percentage_error
score = r2_score(ytest,ypred)
print(score)
```

```python
mape = mean_absolute_percentage_error(ytest,ypred)
mape
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
def poly(degree):
    
    selected_columns = ['surge_multiplier','distance']

    
    

    
    poly_features = PolynomialFeatures(degree=degree)

    
    poly_transform = poly_features.fit_transform(df[selected_columns])

    
    poly_df = pd.DataFrame(poly_transform, columns=poly_features.get_feature_names(selected_columns))

    
    df_poly = pd.concat([df.drop(columns=selected_columns), poly_df], axis=1)


    model = LinearRegression()

    x = df_poly.drop('price',axis=1)
    y = df_poly['price']
    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42015)

    model.fit(xtrain,ytrain)
    ypred = model.predict(xtest)

    from sklearn.metrics import r2_score,mean_absolute_percentage_error
    score = r2_score(ytest,ypred)
    mape = mean_absolute_percentage_error(ytest,ypred)

    return score,mape

```

```python
for i in range(1,10):
    print(i,"degree")
    print(poly(i))
```
